```python
from datasets import Dataset
train_dataset = Dataset.from_pandas(df)
```

```python
def tokenize_function(example):
  
    encoder_max_len = 42 
    decoder_max_len = 42 
    
    toxic = example['en_toxic_comment']
    non_toxic = example['en_neutral_comment']
#     toxic = example['comment_text']
#     non_toxic = example['clean_text']
    # answer = example['answers']['text']
    
#     for 
#     question_plus = f"toxic: {str(toxic)}</s>"
#     answer_plus = f"non-toxic: {str(non_toxic)}</s>"
    encoder_inputs = tokenizer([f"toxic: {str(txt)}" for txt in toxic], padding='max_length', 
                               truncation=True, max_length=encoder_max_len)
    
    decoder_inputs = tokenizer([f"non-toxic: {str(txt)}" for txt in non_toxic], padding='max_length', 
                               truncation=True, max_length=decoder_max_len)
    
    input_ids = encoder_inputs['input_ids']
    input_attention = encoder_inputs['attention_mask']
    target_ids = decoder_inputs['input_ids']
    target_attention = decoder_inputs['attention_mask']
    
#     target_ids[target_ids == tokenizer.pad_token_id] = -100
#     outputs = {'input_ids':input_ids, 'attention_mask': input_attention, 
#                'decoder_input_ids':target_ids, 'decoder_attention_mask':target_attention}
    outputs = {'input_ids':input_ids, 'labels': target_ids}
    return outputs

 
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})    
tokenized_datasets  = train_dataset.map(tokenize_function, batched=True, drop_last_batch=True)
tokenized_datasets.set_format("torch")
# tokenized_datasets = tokenized_datasets.remove_columns(["id", "comment_text", "clean_text"])
tokenized_datasets = tokenized_datasets.remove_columns(["en_toxic_comment", "en_neutral_comment"])
```

```python
# train-test split
new_split_datasets = tokenized_datasets.train_test_split(test_size=0.5, shuffle=True, seed=43)
small_train_dataset, small_eval_dataset = new_split_datasets['train'], new_split_datasets['test']
```

```python
import torch

torch.cuda.empty_cache()
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(small_train_dataset, batch_size=8)
eval_dataloader = DataLoader(small_eval_dataset, batch_size=8)
```

```python
for inp in train_dataloader.dataset:
    print(inp['input_ids'].shape, inp['labels'].shape)
```

```python
from torch.optim import AdamW

optimizer = AdamW(model.parameters(), lr=5e-5)
```

```python
from transformers import get_scheduler

num_epochs = 30
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler(
    name="linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps
)
```

```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model.to(device)
print(device)
```

```python
from tqdm.auto import tqdm

progress_bar = tqdm(range(num_training_steps))

model.train()
for epoch in range(num_epochs):
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)
```

```python
test = list(eval_dataloader)
```

```python
# inference
for i in test:
# input_ids = test[i]['input_ids']
    outputs = model.generate(i['input_ids'].to(device))
    print("[inputs]: ", tokenizer.decode(i['input_ids'][0], skip_special_tokens=True),
        "\n[PRED]: ", tokenizer.decode(outputs[0], skip_special_tokens=True),
         "\n[REAL]: ", tokenizer.decode(i['labels'][0], skip_special_tokens=True))
```
